```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)

import matplotlib.pyplot as plt
import seaborn as sns

import datetime

import warnings
warnings.filterwarnings('ignore')
```

```python
dfTrain = pd.read_csv('train.csv')
dfTest = pd.read_csv('test.csv')
dfTrain.head(3)
```

```python
# checking the null values in dataset train
for i in dfTrain.columns :
    nul = dfTrain[i].isnull().sum()
    if nul != 0 :
        print(f'Column {i} have {nul} null values')

# checking the null values in dataset test
for i in dfTest.columns :
    nul = dfTest[i].isnull().sum()
    if nul != 0 :
        print(f'Column {i} have {nul} null values')
```

```python
dfTrain.describe().T
```

```python
dfTrain.groupby(['Cover_Type']).size()
```

```python
dfTrain = dfTrain[dfTrain['Cover_Type'] != 5]
dfTrain.drop(columns = ['Id', 'Soil_Type7', 'Soil_Type15'], inplace = True)
test = dfTest.drop(columns = ['Id', 'Soil_Type7', 'Soil_Type15'])
```

```python
dfTrain.head(3)
```

```python
dfTrain.columns
```

```python
from tensorflow.keras.utils import to_categorical

X = dfTrain.drop('Cover_Type', axis = 1)
y = to_categorical(dfTrain['Cover_Type'])
X_test = test
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X)
X_test = sc.transform(X_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, BatchNormalization

model = Sequential()

model.add(Dense(52, kernel_initializer = 'normal', activation = 'relu', input_shape = (52, )))

model.add(Dropout(rate = 0.1))
model.add(BatchNormalization())

model.add(Dense(48, kernel_initializer = 'normal', activation = 'relu'))

model.add(Dropout(rate = 0.1))
model.add(BatchNormalization())

model.add(Dense(8, kernel_initializer = 'normal', activation = 'softmax'))

model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])

model_history = model.fit(X_train, y, epochs = 100, batch_size = 64)
```

```python
y_pred = model.predict(X_test)

pred = list()
for i in range(len(y_pred)) :
    pred.append(np.argmax(y_pred[i]))
```

```python
submission = pd.DataFrame({'Id' : dfTest['Id'], 'Cover_Type' : pred})
submission.to_csv('CNN2.csv', index = False, header = True)
```
